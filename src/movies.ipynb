```python
import pandas as pd 
import sqlite3 as sql
```

**Import datasets**

```python
movies = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_movies.csv')
credits = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_credits.csv')
```

```python
print(movies.head(), 'shape', movies.shape)
```

```python
print(credits.head(), 'shape', credits.shape)
```

**Create database SQL**

```python
conn = sql.connect('../data/processed/movies_browser.db')
cursor = conn.cursor()

movies.to_sql('movies', conn, if_exists= 'replace', index= False)
credits.to_sql('credits', conn, if_exists= 'replace', index= False)
```

```python
for row in cursor.execute('SELECT * FROM movies LIMIT 5'):
    print(row)
```

**Merge tables**

```python
query = '''
    SELECT *
    FROM movies
    INNER JOIN credits
    ON movies.title = credits.title;
'''

raw_movies = pd.read_sql_query(query, conn)
conn.close()

raw_movies = raw_movies.loc[:, ~raw_movies.columns.duplicated()]
raw_movies.head()
```

```python
raw_movies.info()
```

```python
movies_clean = raw_movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']].copy()
movies_clean.head()
```

**Transform SQL to json**

```python
import json

def load_json_safe(json_str, default_value = None):
    try:
        return json.loads(json_str)
    except (TypeError, json.JSONDecodeError):
        return default_value

movies_clean['genres'] = movies_clean['genres'].apply(lambda x: [item['name'] for item in json.loads(x)] if pd.notna(x) else None)

movies_clean['keywords'] = movies_clean['keywords'].apply(lambda x: [item['name'] for item in json.loads(x)] if pd.notna(x) else None)

movies_clean['cast'] = movies_clean['cast'].apply(lambda x: [item['name'] for item in json.loads(x)] [:3] if pd.notna(x) else None)

movies_clean['crew'] = movies_clean['crew'].apply(lambda x: ''.join([crew_member['name'] for crew_member in load_json_safe(x) if crew_member['job'] == 'Director']))

movies_clean['overview'] = movies_clean['overview'].apply(lambda x: [x])

movies_clean.head()
```

```python
movies_clean["overview"] = movies_clean["overview"].apply(lambda x: [str(x)])

for var in ['genres','keywords', 'cast', 'crew']:
    movies_clean[var]=movies_clean[var].apply(lambda x: [str(y) for y in x])

movies_clean["tags"] = movies_clean["overview"] + movies_clean["genres"] + movies_clean["keywords"] + movies_clean["cast"] + movies_clean["crew"]
movies_clean["tags"] = movies_clean["tags"].apply(lambda x: ",".join(x).replace(",", " "))

movies_clean.head()
```

```python
movies_clean['overview'] = movies_clean ['overview'].apply(lambda x: [str(x)])

for i in ['genres', 'keywords', 'cast', 'crew']:
    movies_clean[i] = movies_clean[i].apply(lambda x: [str(y) for y in x])

movies_clean['tags'] = movies_clean['overview'] + movies_clean['genres'] + movies_clean['keywords'] + movies_clean['cast'] + movies_clean['crew']
movies_clean['tags'] = movies_clean['tags'].apply(lambda x: ','.join(x).replace(',', ' '))

movies_clean.head()
```

```python
movies_clean.drop(columns= ['genres', 'keywords', 'cast', 'crew', 'overview'], inplace= True)
movies_clean.iloc[0].tags
```

```python
movies_clean.to_csv('../data/processed/movies_clean.csv', index= False)
conn = sql.connect('../data/processed/movies_browser.db')

movies.to_sql('clean_movies_data', conn, if_exists= 'replace', index= False)
conn.close()
```
